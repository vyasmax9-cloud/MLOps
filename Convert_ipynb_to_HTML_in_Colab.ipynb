**Instruction**:    
1. Download the ipynb, which you want to convert, on your local computer.    
2. Run the code below to upload the ipynb.  
3. The html version will be downloaded automatically on your local machine.

Enjoy it!

```python
#@title Convert ipynb to HTML in Colab
# Upload ipynb
from google.colab import files
f = files.upload()

# Convert ipynb to html
import subprocess
file0 = list(f.keys())[0]
_ = subprocess.run(["pip", "install", "nbconvert"])
# Install ipywidgets to handle widget metadata during conversion
_ = subprocess.run(["pip", "install", "ipywidgets"])

# Step 1: Clear outputs of the notebook to resolve KeyError: 'state' issues
print(f"Clearing outputs for {file0}...")
clear_output_result = subprocess.run(
    ["jupyter", "nbconvert", "--clear-output", "--inplace", file0],
    capture_output=True,
    text=True
)
print("Clear output stdout:", clear_output_result.stdout)
print("Clear output stderr:", clear_output_result.stderr)
if clear_output_result.returncode != 0:
    raise subprocess.CalledProcessError(
        clear_output_result.returncode,
        clear_output_result.args,
        output=clear_output_result.stdout,
        stderr=clear_output_result.stderr
    )

# Step 2: Convert the cleared notebook to HTML
print(f"Converting {file0} to HTML...")
nbconvert_result = subprocess.run(
    ["jupyter", "nbconvert", file0, "--to", "html"],
    capture_output=True,
    text=True
)

print("nbconvert stdout:", nbconvert_result.stdout)
print("nbconvert stderr:", nbconvert_result.stderr)

# Check if nbconvert was successful and raise an error if not
if nbconvert_result.returncode != 0:
    raise subprocess.CalledProcessError(
        nbconvert_result.returncode,
        nbconvert_result.args,
        output=nbconvert_result.stdout,
        stderr=nbconvert_result.stderr
    )

# download the html
files.download(file0[:-5]+"html")
```
